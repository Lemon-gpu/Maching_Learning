```python
%matplotlib inline
import torch
from torch.distributions import multinomial 
import tools
```

```python
fair_probs = torch.ones([6]) / 6
multinomial.Multinomial(10, fair_probs).sample() / 10
```

```python
counts = multinomial.Multinomial(10, fair_probs).sample((500,))
cum_counts = counts.cumsum(dim = 0)
temp = cum_counts.sum(dim=1, keepdim=True)
estimate = cum_counts / temp
```

```python
tools.set_figsize((6, 4.5))
for i in range(0, 6):
    tools.plt.plot(estimate[:, i], label=("P(" + str(i+1) + ")"))
tools.plt.axhline(y = 0.167)
tools.plt.show()
```
