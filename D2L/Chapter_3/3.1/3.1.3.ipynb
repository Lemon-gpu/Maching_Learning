```python
%matplotlib inline
import math
import time
import numpy as np
import torch
import tools.Timer as Timer
import tools.plot as plot
```

```python
def normal(x: np.ndarray, mu: float, sigma: float):
    p = 1 / math.sqrt(2 * math.pi * sigma ** 2)
    return p * np.ma.exp(-0.5 / sigma ** 2 * ((x - mu) ** 2))
```

```python
x = np.arange(-10, 10, 0.01)
params = [(0, 1), (0, 2), (3, 1)]

plot.plot(x, [normal(x, mu, sigma) for mu, sigma in params], figsize=(16, 9))
```
